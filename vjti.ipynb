```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV
```

```python
# List of classification models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "Naive Bayes": GaussianNB(),

}
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.2, random_state=42)
```

```python
initial_results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    initial_results[name] = accuracy
    print(f"Before Tuning - {name}: {accuracy:.4f}")
```

```python
# Hyperparameter tuning
hyperparameters = {
    "Logistic Regression": {"C": [0.1, 1, 10]},
    "KNN": {"n_neighbors": [3, 5, 7]},
    "SVM": {"C": [0.1, 1, 10], "kernel": ["linear", "rbf"]},
    "Decision Tree": {"max_depth": [None, 5, 10]},
    "Random Forest": {"n_estimators": [50, 100, 200]},
    "Gradient Boosting": {"learning_rate": [0.01, 0.1, 0.2]},
    "Naive Bayes": {},  }
```

# Train and evaluate models after tuning


```python
tuned_results = {}
for name, model in models.items():
    param_grid = hyperparameters.get(name, {})
    if param_grid:
        grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
        grid_search.fit(X_train, y_train)
        best_model = grid_search.best_estimator_
    else:
        best_model = model.fit(X_train, y_train)

    y_pred = best_model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    tuned_results[name] = accuracy
    print(f"After Tuning - {name}: {accuracy:.5f}")
```

# Display best models before and after tuning


```python
best_initial_model = max(initial_results, key=initial_results.get)
best_tuned_model = max(tuned_results, key=tuned_results.get)
print(f"Best Model Before Tuning: {best_initial_model} with accuracy {initial_results[best_initial_model]:.4f}")
print(f"Best Model After Tuning: {best_tuned_model} with accuracy {tuned_results[best_tuned_model]:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression



# Store the Logistic Regression model in a variable
logistic_regression_model = models["Logistic Regression"]


X = clean_data[scaled_X.columns]
y = clean_data['is_healthy']


x_min, x_max = X['phy_fitness'].min() - 1, X['phy_fitness'].max() + 1
y_min, y_max = X['mindfulness'].min() - 1, X['mindfulness'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))


X_grid = pd.DataFrame(np.c_[xx.ravel(), yy.ravel()], columns=['phy_fitness', 'mindfulness'])


for feature in scaled_X.columns:
    if feature not in ['phy_fitness', 'mindfulness']:
        X_grid[feature] = X[feature].mean()

# Reorder columns to match the order used during training
X_grid = X_grid[scaled_X.columns]


Z = logistic_regression_model.predict(X_grid)
Z = Z.reshape(xx.shape)

# Plot the decision boundary and data points
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X['phy_fitness'], X['mindfulness'], c=y, cmap=plt.cm.Paired)
plt.xlabel('phy_fitness')
plt.ylabel('mindfulness')
plt.title('Decision Boundary (Logistic Regression)')
plt.show()
```
